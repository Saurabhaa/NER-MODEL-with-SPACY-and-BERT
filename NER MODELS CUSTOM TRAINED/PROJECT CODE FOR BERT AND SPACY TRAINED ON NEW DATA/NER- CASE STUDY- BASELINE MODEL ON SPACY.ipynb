```python
# Missing values
df['Sentence #'] = df['Sentence #'].fillna(method='ffill')
df['Word'] = df['Word'].fillna('')
```

```python
# Taking sentences and their ner tags
sentences = df.groupby('Sentence #')['Word'].apply(list).values
ner_tags = df.groupby('Sentence #')['Tag'].apply(list).values
```

```python
 # Making sure all elements are string
    
sentences = [[str(word) if not isinstance(word, str) else word for word in sentence] for sentence in sentences]
```

```python
# identifiying spaces before punctuation and removal --
# remove extra spaces and tabs
sentences=[re.sub(r'\s+([?.!"])', r'\1', sentence) for sentence in sentences]
sentences = [re.sub(r'\s+', ' ', sentence) for sentence in sentences]
sentences
```

```python
# en core load
nlp = spacy.load("en_core_web_sm")

# Data annotation....
#check valid spans
docs = []
for sentence, tags in zip(sentences, ner_tags):
    doc = nlp.make_doc(sentence)
    ents = []
    start = 0
    for word, tag in zip(sentence.split(), tags):
        if tag != 'O':
            start = sentence.find(word, start)
            end = start + len(word)
            span = doc.char_span(start, end, label=tag)
            if span is not None:
                ents.append(span)
            else:
                print(f"Warning: span is None for word '{word}' in sentence '{sentence}'")
            start = end
    doc.ents = ents
    docs.append(doc)

```

```python

# Split the data
train_docs, temp_docs = train_test_split(docs, test_size=0.2)
val_docs, test_docs = train_test_split(temp_docs, test_size=0.5)

# Create DocBin objects
train_docbin = DocBin(docs=train_docs)
val_docbin = DocBin(docs=val_docs)
test_docbin = DocBin(docs=test_docs)

# Save to disk
train_docbin.to_disk("./train.spacy")
val_docbin.to_disk("./val.spacy")
test_docbin.to_disk("./test.spacy")
```

```python
# ENTS_F (Entity F1-score):
# This metric represents the overall performance of the named entity recognition (NER) model.
# It combines precision (ENTS_P) and recall (ENTS_R) into a single value.
# Higher ENTS_F indicates better NER performance.
# ENTS_P (Entity Precision):
# Precision measures the proportion of correctly predicted named entities out of all predicted entities.
# It answers the question: “Of the entities predicted by the model, how many were actually correct?”
# ENTS_R (Entity Recall):
# Recall measures the proportion of correctly predicted named entities out of all true entities in the data.
# It answers the question: “Of all the actual entities, how many did the model manage to predict?”
```

```python
# Baseline model evaluation...

# Load the trained model
nlp_trained = spacy.load("./output/model-best")

# Evaluate on the test set
scorer = Scorer()
examples = []
for doc in test_docs:
    example = Example.from_dict(doc, {"entities": [(ent.start_char, ent.end_char, ent.label_) for ent in doc.ents]})
    examples.append(example)
scores =scorer.score(examples)
print(scores)
```

```python
# Define a function to predict NER tags for a given sentence
def predict_ner(sentence):
    """
    Predict NER tags for each word in the given sentence.
    """
    doc = nlp_trained(sentence)
    predictions = [(token.text, token.ent_iob_, token.ent_type_) for token in doc]
    return predictions

# Example sentence
sentence = "John Doe went to New York last week."

# Predict NER tags
predictions = predict_ner(sentence)

# Print the predictions
for word, iob, ner in predictions:
    print(f"Word: {word}, IOB: {iob}, NER: {ner}")
```
